# $t\bar{t}$-analysis
The chapter is covering the $t\bar{t}$-analysis logic. [2015 CMS Open Data](https://cms.cern/news/first-cms-open-data-lhc-run-2-released) is assumed as input data for concreteness. The basic idea is to give a basic algorithm for distinguishing t-quark pair production channel among other concurrent channels (production of single t-quark and w-jet) and to get the mass peak of t-quark using a histogram.

## Input
Input data is a set of five samples of root-files. Each sample is produced in MC simulation and represents partial interaction channel, one of five: $t\bar{t}$-channel, single top s-channel, single top t-channel, single top tW-channel, Wjets-channel.

## Output
The analysis task is a selection events from whole set, in which measured quantities were originated from $t\bar{t}$-decay. In real data one can not exatly know in which event $t\bar{t}$-pair were produced. But since we have simulated data, we know exactly whether or not $t\bar{t}$ in any particular event. That is defined by set to which file belongs, as was mentioned above. As five channels were involved, five different sets were generated and are given as input data ('ttbar', 'single_top_s_chan', 'single_top_t_chan', 'single_top_tW', 'wjets').
To select events we will apply some criteria (explained below) and then compare relative rate of those events which indeed were generated with $t\bar{t}$-quark pair production. The example below demonstrate successfully performed $t\bar{t}$ analysis. That is concludet from the fact, that most of the selected events indeed belong to $t\bar{t}$-channel.



## Events filtering algorithm
Not all particles with it's quantities are needed for doing analysis. The only ones that is needed to applying some selection criterias are leptons (electrons and muons) and jets that is the products of $t\bar{t}$ decay. Since semi-leptonic decay is dominant process, two jets, two b-jets and one outgoing lepton are expected to be detected, as can be concluded from the diagram below:

The algorithm how to select only appropriative events is presentet below:

### 1. Filtering events by transverse momentum value
The transverse momentum of electrons, muons and jets must be higher then some threshold. It is well suggestion to select threshold value 25 GeV. In this step one have all simuleted events as input data. The output is only those events which contains at least one electron, muon and jet with transverse momentum higher then 25 GeV.

### 2.  Filtering events by lepton and jet numbers
Further, events that contain only one lepton (exactly one electron or exactly one muon) and four or more jets must be selected.

### 3. Filtering events by jet b-tagging
In the next, events that contain at least two b-tagget jets must be selected. It is well suggestion to put b-tag threshold equal to 0.5. So, the output is only those events which contains at least two jets with b-tag value higher than 0.5.

## Trijet mass plotting algorithm
For this moment one had to leave only appropriate events for our analysis. This is close to final goal: plotting trijet mass.
The task is to find all possible trijet combinations per event and select the most appropriative one. To each event we have to assign trijet which is assumed to be the product of $t\bar{t}$ decay.\
So to finish analysis it is needed:
* Build all possible three-jet combinations per event.
* Find the maximum b-tag value for each jet in each combination.
* Leave only those combination with corresponding b-tag value more than selected threshold(>0.5).
* Select trijet with maximum transverse momentum.
As output one has the sample of events with assigned trijet for each events.

At this point all is ready to plot trijet mass.
